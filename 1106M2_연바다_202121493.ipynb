# 결측데이터 처리하기

## 결측데이터 확인

```python
# 데이터 불러오기
import pandas as pd
filename='/content/sample_data/train.csv'
df=pd.read_csv(filename)
df.head() #상위 5행 출력 #NaN=데이터 프레임 내에 데이터가 없는 결측 데이터 의미함.
```

```python
# 결측데이터('Embarked') 확인하기
df[df['Embarked'].isna()] #'Embarked'컬럼의 결측 데이터 출력
```

```python
# 결측데이터('Age') 확인하기
#자료가 많으므로 .head() 설정
df[df['Age'].isna()].head() #Age컬럼에는 너무 많은 결측데이터 존재=>head로 제어(5개만 출력)
```

```python
# 연습 - 결측데이터('Cabin') 확인하기
#자료가 많으므로 .head() 설정 #자료가 너무 많아 너무 많은 행이 출력되어 코랩이 멍~해짐
df[df['Cabin'].isna()].head()
```

```python
# 컬럼별 결측데이터 개수 확인
df.isnull().sum() #데이터 프레임 전체의 각 컬럼별 결측데이터 개수
```

## 결측데이터 삭제

```python
#drop 함수 이용
#Cabin 컬럼 전체 삭제

df.drop('Cabin',axis=1,inplace=True)
```

```python
#컬럼 이름 확인
df.columns #위의 코딩으로 인해 Cabin컬럼 없어짐
```

```python
#삭제된 데이터 프레임 확인

df.head()
```

## 결측데이터 대체 : 최빈값으로 채우기

```python
# 최빈값 찾기
# 'Embarked' 컬럼에 가장 빈번한 값은 S이고, S를 인덱싱 0으로 지정.
df['Embarked'].mode() #결측값을 최빈값으로 대체하기 위해 최빈값을 출력
                      #결과에서 0이 인덱스 번호임
```

```python
#Embarked 컬럼의 결측데이터를 인덱싱 0 (값S로 대체)

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True) #결측 데이터가 2개뿐이므로 이 컬럼 자체를 삭제하는 것이 아니라 결측값 2개를 최빈값으로 대체
df.head()
```

```python
#Embarked 컬럼의 결측데이터 확인
df[df['Embarked'].isna()] #Embarked컬럼에서 이제 결측데이터 출력 안됨
```

```python
# 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인
df.isnull().sum() #결측데이터가 없다고 뜸(대체 되었기 때문)
```

```python
# 결측데이터 대체되기 전 결측데이터 포함 행들 출력 및 확인
df.loc[61, 'Embarked']
```

```python
df.loc[829, 'Embarked']
```

```python
# 최빈값 찾기
# 'Age' 컬럼에 가장 빈번한 값은 S이고, S를 인덱싱 0으로 지정.
df['Age'].mode()
```

```python
#연습 - Age 컬럼의 결측데이터를 인덱싱 0 값S로 대체
#상위 10행을 출력하여, 인덱싱 5의 결측데이터 대체상태 확인 #head(10)으로 출력
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
df.head(10)
```

```python
#연습 - Age 컬럼의 결측데이터 확인
df[df['Age'].isna()]
```

```python
# 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인
df.isnull().sum()
```

```python
# 결측데이터 대체되기 전 결측데이터 포함 행들 출력 및 확인
#5,17,19 인덱싱  #df.loc[인덱싱, 컬럼이름]
df.loc[5, 'Age']
```
